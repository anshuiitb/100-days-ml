<h2>Line Plot</h2>

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights.groupby('year').sum('passengers')
```

```python
flights.groupby('year')['passengers'].sum()
```

```python
flights.groupby('year').agg({
    'passengers': ['sum', 'mean', 'max', 'min']
})
```

```python
flights.groupby('year').agg(
    total_passengers = ('passengers', 'sum'),
    avg_passengers   = ('passengers', 'mean')
)
```

```python
titanic.groupby('Pclass').agg({
    'Age': 'mean',
    'Fare': ['mean', 'median'],
    'Survived': 'mean'
})
```

```python
def range_func(x):
    return x['passengers'].max() - x['passengers'].min()

flights.groupby('year').apply(range_func)
```

```python
def zscore(group):
    return (group - group.mean()) / group.std()

flights.groupby('year')['passengers'].apply(zscore)
```

```python
flights['yearly_total'] = flights.groupby('year')['passengers'].transform('sum')
flights['yearly_pct'] = flights['passengers'] / flights['yearly_total']*100
```

```python
flights.tail()
```

```python
def drop_columns(df):
    return df.drop(['Cabin', 'Ticket'], axis=1)

def fill_age(df):
    df['Age'] = df['Age'].fillna(df['Age'].median())
    return df

titanic_clean = (
    titanic
    .pipe(drop_columns)
    .pipe(fill_age)
    .assign(AgeSquared = lambda df: df['Age']**2)
)
```
