```python
import requests
import pandas as pd
import time
from tqdm import tqdm
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

# --- Setup a session with retry & backoff ---
session = requests.Session()
retries = Retry(
    total=5,                     # retry up to 5 times
    backoff_factor=1,            # 1s, 2s, 4s, ...
    status_forcelist=[429, 500, 502, 503, 504],
    allowed_methods=["GET"]
)
session.mount("https://", HTTPAdapter(max_retries=retries))

# --- API setup ---
api_key = "c06b1614682dfc451e035ec531661fe6"
base_url = "https://api.themoviedb.org/3/movie/top_rated"

# --- Create empty DataFrame once ---
df = pd.DataFrame()

# --- Loop through pages safely ---
for i in tqdm(range(1, 525)):   # TMDB only has ~500 pages max
    try:
        url = f"{base_url}?api_key={api_key}&language=en-US&page={i}"
        response = session.get(url, timeout=15)
        response.raise_for_status()  # catch 4xx / 5xx
        json_data = response.json()

        if 'results' not in json_data:
            print(f"⚠️ Skipping page {i}: no results found.")
            continue

        temp_df = pd.DataFrame(json_data['results'])[
            ['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']
        ]
        df = pd.concat([df, temp_df], ignore_index=True)

        time.sleep(0.2)  # small delay to avoid API throttling
    except requests.exceptions.SSLError as e:
        print(f"🔁 SSL error on page {i}, retrying after 3s...")
        time.sleep(3)
        continue
    except Exception as e:
        print(f"❌ Error on page {i}: {e}")
        time.sleep(2)
        continue

print(f"✅ Done! Total movies fetched: {len(df)}")
df.head()
```

```python
import requests
import pandas as pd
import time
from tqdm import tqdm  # optional progress bar

# --- RapidAPI details ---
RAPIDAPI_KEY = "00b03c4caemsh7d1637fa7f55447p11bc21jsn39418b228642"
RAPIDAPI_HOST = "yahoo-finance15.p.rapidapi.com"
BASE_URL = f"https://{RAPIDAPI_HOST}/api/v1/markets/insider-trades"

# --- Headers ---
headers = {
    "x-rapidapi-key": RAPIDAPI_KEY,
    "x-rapidapi-host": RAPIDAPI_HOST,
}

# --- Storage ---
all_records = []
page = 1
total_pages = 13311  # (from your output)
per_page = 100

# --- Loop over pages ---
for page in tqdm(range(1, total_pages + 1), desc="Fetching Insider Trades"):
    params = {"page": page}   # ✅ API uses ?page= parameter
    response = requests.get(BASE_URL, headers=headers, params=params)

    # handle API errors
    if response.status_code != 200:
        print(f"⚠️ Skipping page {page} — Status {response.status_code}")
        time.sleep(2)
        continue

    data = response.json()
    body = data.get("body", [])
    if not body:
        break  # stop if no more data

    all_records.extend(body)

    # be nice to the API (avoid rate limiting)
    time.sleep(0.2)

# --- Convert to DataFrame ---
df = pd.DataFrame(all_records)
df.to_csv("yahoo_insider_trades.csv", index=False)
print(f"✅ Done! Saved {len(df)} records from {page} pages.")
```
