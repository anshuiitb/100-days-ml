```python
from bs4 import BeautifulSoup
import requests

url = "https://www.ambitionbox.com/list-of-companies?page=1"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}

r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, "lxml")

# Loop over each company card
for card in soup.select(".companyCardWrapper"):
    name = card.find("h2").get_text(strip=True) if card.find("h2") else "Unknown"

    # Default values
    high_val, crit_val = "none", "none"

    # Look for "Highly Rated For"
    high_header = card.select_one(".companyCardWrapper__ratingHeader--high")
    if high_header:
        val_tag = high_header.find_next("span", class_="companyCardWrapper__ratingValues")
        if val_tag:
            high_val = val_tag.get_text(strip=True)

    # Look for "Critically Rated For"
    crit_header = card.select_one(".companyCardWrapper__ratingHeader--critical")
    if crit_header:
        val_tag = crit_header.find_next("span", class_="companyCardWrapper__ratingValues")
        if val_tag:
            crit_val = val_tag.get_text(strip=True)

    print(f"{name:25s} | Highly Rated For: {high_val:60s} | Critically Rated For: {crit_val}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

url = "https://gymkhana.iitb.ac.in/~sustainabilitycell/#about"

options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

driver.get(url)
time.sleep(3)   # wait for React to load and render

html = driver.page_source
soup = BeautifulSoup(html, "lxml")

print(soup.prettify())
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

url ="https://gesh.iitb.ac.in/aboutus"

options = Options()
options.add_argument("--headless=new")   # better headless mode
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--no-sandbox")
options.add_argument("--disable-gpu")

driver = webdriver.Chrome(options=options)
driver.get(url)

# Wait for React content to load (wait for any element that appears only after rendering)
try:
    WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div, section, p"))
    )
except:
    print("Timeout — React content didn't load.")

html = driver.page_source
soup = BeautifulSoup(html, "lxml")
print(soup.prettify())

driver.quit()
```

```python
for text in soup.find_all("fadeInUp"):
    print(text.text.strip())
```
